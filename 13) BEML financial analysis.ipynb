### To selcet respective columns of date and close

```python
data1=data[["Date","Close"]]
data1
```

```python
data1.describe()
```

```python
data1.plot(y=["Close"])
plt.title("close trend")
plt.show()
```

```python
data1["Gain"]=data1.Close.pct_change(periods=1)
```

```python
data1
```

```python
data1=data1.dropna()   # dropna : drop all null values
data1
```

```python
data1=data1.set_index(pd.DatetimeIndex(data1['Date']))
data1
```

```python
plt.figure(figsize=(8,6))
plt.plot(data1.index, data1.Gain)
plt.xlabel=('Time')
plt.ylabel=('Gain')
```

```python
round(data1.Gain.mean(),4)
```

```python
round(data1.Gain.std(),4)
```
